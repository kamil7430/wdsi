Przygotowanie danych do regresji

```python
import pandas as pd
# data = pd.read_csv("data/regresja/domy.csv").dropna(how="any")
# data = pd.read_csv("data/regresja/domy.csv").fillna(value=0)
data = pd.read_csv("data/regresja/domy.csv")
data = pd.get_dummies(data, dtype=int)
data = data.interpolate() # Interpoluje NAN-y a nie '?'
# data = data.bfill()

y = data["SalePrice"]
X = data.drop("SalePrice", axis="columns")
```

Coś tam

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, accuracy_score

reg = linear_model.LinearRegression()
_ = reg.fit(X, y)

y_pred = reg.predict(X)
mean_squared_error(y, y_pred)
```

Coś innego

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)
reg = linear_model.LinearRegression()
_ = reg.fit(X_train, y_train)

reg.score(X_test, y_test)

y_pred = reg.predict(X_test)
```

```python
import matplotlib.pyplot as plt
plt.scatter(y_test, y_pred, color="red")
plt.plot([0, 500_000], [0, 500_000], color="blue")
plt.show()
```
